```python
import getpass
import os

from langchain.document_loaders import TextLoader, DirectoryLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
```

```python
# %cd drive/MyDrive/CISCO/Assignment/task_2
%cd task_2
!ls
```

```python
loader = DirectoryLoader('./products/', glob="**/*.txt", show_progress=True)
documents = loader.load()
```

```python
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
db = FAISS.from_documents(docs, embeddings)
```

```python
db.save_local("faiss_index")
```

```python
query = "What type of Cisco Switches are available?"
K = 10
```

```python
results_with_scores = db.similarity_search_with_score(query, k=K)
```

```python
for doc, score in results_with_scores:
    print(f"Content: {doc.page_content}\nMetadata: {doc.metadata}\nScore: {score}")
    print('\n\n')
```
